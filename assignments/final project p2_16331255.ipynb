```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('/Users/lucykinnear/Documents/GitHub/POP77001_Computer_Programming_2021/data/listings.csv')
```

```python
plt.scatter(data['longitude'],data['latitude'])
plt.show()
```

```python
location = data[['longitude', 'latitude']].values
```

```python
kmeans3 = KMeans(3)
kmeans3.fit(location)
```

```python
identified_clusters3 = kmeans3.fit_predict(location)
identified_clusters3
```

```python
data_with_clusters3 = data.copy()
data_with_clusters3['Clusters'] = identified_clusters3 
plt.scatter(data_with_clusters3['longitude'],data_with_clusters3['latitude'],c=data_with_clusters3['Clusters'],cmap='rainbow')
```

```python
# phoenix park area, foxrock/dun laoghaire, malahide
```

```python
kmeans5 = KMeans(5)
kmeans5.fit(location)
```

```python
identified_clusters5 = kmeans5.fit_predict(location)
identified_clusters5
```

```python
data_with_clusters5 = data.copy()
data_with_clusters5['Clusters'] = identified_clusters5
plt.scatter(data_with_clusters5['longitude'],data_with_clusters5['latitude'],c=data_with_clusters5['Clusters'],cmap='rainbow')
```

```python
# blanch, phoenix park, east wall, deansgrange/dun laoghaire, malahide
```
